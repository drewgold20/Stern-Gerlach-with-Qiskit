```python
#importing Aer Simulator 
from qiskit_aer import AerSimulator
```

```python
# importing other packages in order to run and view our circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from qiskit import transpile
```

```python
from qiskit.visualization import plot_histogram
```

```python
# creating a quantum state
import random
from numpy import pi
import numpy as np

# remember that a and b need to add up to 1 after we sqare and take absolute value of each 
a = 1/np.sqrt(2)
b = 1/np.sqrt(2)
# normalizing a and b. if it is not normalized our results do not mean anything
norm = np.sqrt(a*np.conjugate(a)+b*np.conjugate(b))
a = a/norm
b = b/norm
print(a,b)
```

```python
# building and printing our quantum circuit
qr = QuantumRegister(1, 'q')
cr = ClassicalRegister(1, 'c')
qc = QuantumCircuit(qr, cr)

# Initialize the quantum state
qc.initialize([a, b])
qc.measure(0,0)
qc.draw('mpl')
```

```python
# Transpile for simulator
simulator = AerSimulator()
qc = transpile(qc, simulator)

# Run and get counts
result = simulator.run(qc).result()
counts = result.get_counts(qc)
plot_histogram(counts, title='Probabilities ')
```
